### Data Preprocessing
* Lower case
* Tokenization
* Removing Special characters
* Removing stop words and punctuation
* stemming

```python
# stop words
nltk.download('stopwords')
from nltk.corpus import stopwords
stopwords.words('english')
```

```python
import string
string.punctuation
```

```python
# stemming
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def transform_text(text):
    text = text.lower()
    text = nltk.word_tokenize(text)
    t = []
    for i in text:
        if i.isalnum():
            t.append(i)
    text = t[:]
    t.clear()
    
    for i in text:
        if i not in stopwords.words('english') and i not in string.punctuation:
            t.append(i)
    text = t[:]
    t.clear()
    
    for i in text:
        t.append(ps.stem(i))
        
    text = " ".join(t)
    
    return text
```

```python
transform_text('I loved dancing on the streets. I like Machine Learning too. about U?')
```

```python
df['transformed_text'] = df['text'].apply(transform_text)
df.head()
```

```python
spam_corpus = []
for msg in df[df['target']==1]['transformed_text'].tolist():
    for word in msg.split():
        spam_corpus.append(word)

len(spam_corpus)
```

```python
pd.DataFrame(spam_corpus).value_counts()
```

```python
from collections import Counter
Counter(spam_corpus)

cmmn_words_spam = pd.DataFrame(Counter(spam_corpus).most_common(30))
cmmn_words_spam.head(4)
```

```python
cmmn_words_spam.columns = ['word','count']
```

```python
sns.barplot(y =cmmn_words_spam['word'],x=cmmn_words_spam['count'],palette='pastel')
```

```python
ham_corpus = []
for msg in df[df['target']==0]['transformed_text'].tolist():
    for word in msg.split():
        ham_corpus.append(word)

cmmn_words_ham = pd.DataFrame(Counter(ham_corpus).most_common(30))
cmmn_words_ham.columns = ['word','count']
cmmn_words_ham.head(4)
```
